```python
#Import libraires
import pandas as pd
from selenium.webdriver.common.by import By
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
```

```python
#path for chrome webdriver
path=Service("C:\selenium drivers\chromedriver.exe")
op = webdriver.ChromeOptions()
```

#### *Scraping the top 100 cytpocurrency data from Coin Marketcap*

```python
#Initializing list to store the data
coin_name_list=[]
coin_symbol_list=[]
coin_price_list=[]
hr_24_change_list=[]
market_cap_list=[]
volume_24_list=[]

#main code here
try:
    #URL path
    driver = webdriver.Chrome(service=path, options=op)
    # For maximizing window
    driver.maximize_window()
    # Implicit wait apply for all elements to wait for 5 seconds to load
    driver.implicitly_wait(5)
    driver.get("https:/coinmarketcap.com/")
    for i in range(1,101):
        flag=driver.find_element(by=By.XPATH,
                                   value="//*[@id='__next']/div/div[1]/div[2]/div/div/div[5]/table/tbody/tr[" + str(i) + "]/td[3]/div/a/div/div/p")
        
        #Accessing coin name, price,24 hr change, Mkt Cap using the element xpath
        
        coin_name = driver.find_element(by=By.XPATH,
                                   value="//*[@id='__next']/div/div[1]/div[2]/div/div/div[5]/table/tbody/tr[" + str(i) + "]/td[3]/div/a/div/div/p")
        
        coin_symbol=driver.find_element(by=By.XPATH,
                                value="//*[@id='__next']/div/div[1]/div[2]/div/div/div[5]/table/tbody/tr[" + str(i) + "]/td[3]/div/a/div/div/div/p")
            
        coin_price=driver.find_element(by=By.XPATH,
                                        value="//*[@id='__next']/div/div[1]/div[2]/div/div/div[5]/table/tbody/tr["+ str(i) + "]/td[4]/div/a/span")
     
        hr_24=driver.find_element(by=By.XPATH,
                                value="//*[@id='__next']/div/div[1]/div[2]/div/div/div[5]/table/tbody/tr["+ str(i) + "]/td[5]/span")
        
        market_cap=driver.find_element(by=By.XPATH,
                                value="//*[@id='__next']/div/div[1]/div[2]/div/div/div[5]/table/tbody/tr["+ str(i) + "]/td[7]/p/span[2]")
        
        volume_24hr=driver.find_element(by=By.XPATH,
                                value="//*[@id='__next']/div/div[1]/div[2]/div/div/div[5]/table/tbody/tr["+ str(i) + "]/td[8]/div/a/p")
         
        
        driver.execute_script("arguments[0].scrollIntoView();",flag)
        
        #Printing the scraped results
#         print(coin_name.text.replace("\n","").strip(),coin_symbol.text.replace("\n","").strip(),coin_price.text.replace("\n","").strip(),hr_24.text.replace("\n","").strip(),
#              market_cap.text.replace("\n","").strip(),volume_24hr.text.replace("\n","").strip())
        
        #Appending all Data into respective records
        coin_name_list.append(coin_name.text.replace("\n","").strip())
        coin_symbol_list.append(coin_symbol.text.replace("\n","").strip())
        coin_price_list.append(coin_price.text.replace("\n","").strip())
        hr_24_change_list.append(hr_24.text.replace("\n","").strip())
        market_cap_list.append(market_cap.text.replace("\n","").strip())
        volume_24_list.append(volume_24hr.text.replace("\n","").strip())

except:
    #setting time to close after 2 sec
    time.sleep(2)
    driver.close()
finally:
    # setting time to close after 2 sec
    time.sleep(2)
    driver.close()
```

#### *Creating Dataframe from scraped results*

```python
#Create Pandas DataFame
crypto_df=pd.DataFrame({"Name":coin_name_list,"Symbol":coin_symbol_list,"Price":coin_price_list,"24hr Change":hr_24_change_list,"Market Cap":market_cap_list,"Volume_24hr":volume_24_list})
```
